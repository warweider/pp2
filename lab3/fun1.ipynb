```python
def solve(numheads, numlegs):
    if (numlegs - 2 * numheads) % 2 != 0:
        return "No solution number of legs must be even."
    y = (numlegs - 2 * numheads) // 2
    x = numheads - y
    if x < 0 or y < 0:
        return "No solution negative number of animals isn't possible."
    return x, y

numheads = 35
numlegs = 94
result = solve(numheads, numlegs)
print(result)


    
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def filter_prime(numbers):
    num_list = map(int, numbers.split()) 
    return [num for num in num_list if is_prime(num)]

numbers = "2 3 4 5 10 13 17 19 113"
print(filter_prime(numbers))
```

```python
def get_permutations(s, step=0):
    if step == len(s):
        print("".join(s)) 
        return
    
    for i in range(step, len(s)):
        s_list = list(s)
        s_list[step], s_list[i] = s_list[i], s_list[step]
        
        get_permutations(s_list, step + 1)

user_input = input()
get_permutations(list(user_input))
```

```python
def reverse(str):
    words = str.split()
    rstr = " ".join(words[::-1])
    return rstr
a = input()
print(reverse(a))
```

```python
def has_33(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False    
nums = list(map(int,(input("Enter integers: ").split())))
print (has_33(nums))
```

```python
def spy_game(nums):
    code = [0, 0, 7] 
    index = 0  
    for num in nums:
        if num == code[index]:  
            index += 1
        if index == len(code): 
            return True
    
print(spy_game([1,2,4,0,0,7,5]))
print(spy_game([1,0,2,4,0,5,7]))
print(spy_game([1,7,2,0,4,5,0]))
```

```python
import math
def Volume_of_the_sphere(radius):
    return (4/3)*math.pi*pow(radius,3)

radius = float(input("Enter your radius: "))
print(Volume_of_the_sphere(radius))
```

```python
def unique_list(list):
    new_list = []
    for i in list:       
        if i not in new_list:
            new_list.append(i)
    print(new_list)
list = [1, 2, 3, 3, 2, 5, 9, 10, 9]
unique_list(list)
```
