```python
from datetime import datetime, timedelta

current_date = datetime.today()
new_date = current_date - timedelta(days=5)
print("Today:", current_date.strftime("%Y-%m-%d"))
print("Date 5 days before:", new_date.strftime("%Y-%m-%d"))
```

```python
from datetime import datetime, timedelta

today = datetime.today()
yesterday = current_date - timedelta(days=1)
tomorrow = today + timedelta(days = 1)
print("Today:", current_date.strftime("%Y-%m-%d"))
print("Yesterday:", yesterday.strftime("%Y-%m-%d"))
print("Tomorrow:", tomorrow.strftime("%Y-%m-%d"))

```

```python
from datetime import datetime

current_datetime = datetime.now()
formatted_datetime = current_datetime.replace(microsecond=0)
print("Original Datetime:", current_datetime)
print("Datetime without Microseconds:", formatted_datetime)
```

```python
from datetime import datetime

date1 = datetime(2025, 2, 19, 16, 0, 0)  
date2 = datetime(2025, 2, 18, 10, 45, 0) 
difference_in_seconds = abs((date1 - date2).total_seconds())
print(difference_in_seconds)
```
