```python
import re

s = "abbb"
pattern = r'a*b*'
match = re.fullmatch(pattern, s)
print(bool(match))
```

```python
import re

s = "abb"
pattern = r'ab{2,3}'
match = re.fullmatch(pattern, s)
print(bool(match))
```

```python
import re

s = "this_is_a_test example_text more_words"
pattern = r'\b[a-z]+_[a-z]+\b'
matches = re.findall(pattern, s)
print(matches)
```

```python
import re

s = "Hello World Test Example"
pattern = r'\b[A-Z][a-z]+\b'
matches = re.findall(pattern, s)
print(matches)

```

```python
import re

s = "axxxb"
pattern = r'a.*b'
match = re.fullmatch(pattern, s)
print(bool(match))
```

```python
import re

s = "Hello, world. Python regex!"
pattern = r'[ ,.]'
result = re.sub(pattern, ':', s)
print(result)
```

```python
import re

s = "hello_world_example"
pattern = r'_([a-z])'
result = re.sub(pattern, lambda x: x.group(1).upper(), s)
print(result)
```

```python
import re

s = "HelloWorldExample"
pattern = r'(?=[A-Z])'
result = re.split(pattern, s)
print(result)
```

```python
import re

s = "HelloWorldExample"
pattern = r'([a-z])([A-Z])'
result = re.sub(pattern, r'\1 \2', s)
print(result)
```

```python
import re

s = "HelloWorldExample"
pattern = r'([A-Z])'
result = re.sub(pattern, r'_\1', s).lower().lstrip('_')
print(result)
```
