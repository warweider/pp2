```python
def calc(s):
    uppers = sum(map(str.isupper, s))
    lowers = sum(map(str.islower, s))

    return(uppers, lowers)

a = "THISisTeSt"

print(calc(a))
```

```python
def check(p): 
    return p == p[::-1]


a = "madam"
b = "Hello"
print(check(a))
print(check(b))
```

```python
import time
import math

num = int(input())
delay = int(input())

time.sleep(delay / 1000)  

result = math.sqrt(num)

print(f"Square root of {num} after {delay} miliseconds is {result}")
```
